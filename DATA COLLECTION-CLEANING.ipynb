# FOOTBALL RATING
- In this notebook, the data is going to be prepared and cleaned for the next step which is "Feature Engineering". 

- The first thing will be import the libraries.

```python
import pandas as pd
import numpy as np
```

- After that, the data will be taken from differents csv.
- Those "csv"s had been taken from a page using "Data Scraper".
- Because of the difficult to get the data, I make a sample using "Simple Random Sample" technique. Using that, every match date has the same probability to appear.
- In numbers_dataset object there are every number of date.

```python
lista=[]
numbers_dataset=(5,2,8,17,24,25)
for i in numbers_dataset:
    h=f"DATASETS/FECHA-{i}/FECHA-{i}-GENERAL.csv"
    lista.append(h)
general=pd.DataFrame()
for files in lista:
    df=pd.read_csv(files)
    general=pd.concat([general,df])
```

- Now that there is the dataset, it is the time for the cleaning.

```python
## This is only for seeing all the columns from the dataset.
pd.set_option("display.max_columns", None)
```

- For getting a first approach to the data, I will use .info() method.

```python
general.info()
```

- I must eliminate duplicated columns

```python
new_df=general.drop(["JUGADOR","POSICION.1","RATING.1","JUGADOR.1","POSICION.2","RATING.2","JUGADOR.2","POSICION.3","RATING.3","JUGADOR.3","POSICION.4","RATING.4","JUGADOR.4","RATING.5","PASES","TACKLES","DUELOS_TIERRA","DUELOS_AEREOS"],axis=1)
```

```python
new_df.head()
```

```python
new_df.shape
```

- As it can see, there are many columns which have multiple information. Those columns will be splitted.

```python
lista=["DUELOS","GAMBETAS","CROSSES","LONG BALLS","DUELOS_TERRESTRES","DUELOS_AEREOS.1","RUNS_OUT"]
```

```python
columnas=pd.DataFrame()
for i in lista:
    h=new_df[i].str.split('(').apply(pd.Series)
    columnas=pd.concat([columnas,h], axis=1)
columnas.head()
```

- Now, the columns splitted have others names so they are going to be renamed.

```python
names=["DUELOS_R","DUELOS_E","GAMBETAS_R","GAMBETAS_E","CROSSES_R","CROSSES_E","LONG_BALLS_R","LONG_BALLS_E","DUELOS_TERRESTRES_R","DUELOS_TERRESTRES_E","DUELOS_AEREOS_R","DUELOS_AEREROS_E","RUNS_OUT_R","RUN_OUT_E"]
columnas.columns=names
```

```python
columnas.head()
```

- In order to get the data much cleaned as it can, It is going to be changed some things.

```python
for i in columnas.columns:
    columnas[i]=columnas[i].str.replace(')'," ")
columnas
```

- Another split to clean the dataset.

```python
pases0=new_df["ACC_PASSES"].str.split('/',
                                     expand=True)
pases1=pases0[1].str.split('(',
                         expand=True)
pases0.drop(1,inplace=True,axis=1)

```

```python
columnas1=pd.concat([pases0,pases1],axis=1)
```

```python
names=["PASES_R","PASES_E","%PASES_E"]
columnas1.columns=names
```

```python
for i in columnas1.columns:
    columnas1[i]=columnas1[i].str.replace(')'," ")
columnas1["%PASES_E"]=columnas1["%PASES_E"].str.replace("%","")
columnas1
```

- Now that the dataset is clean, the old columns will be drop it.

```python
new_df=new_df.drop(["DUELOS","GAMBETAS","CROSSES","LONG BALLS","DUELOS_TERRESTRES","DUELOS_AEREOS.1","RUNS_OUT","ACC_PASSES"],
            axis=1)
```

- The "rating" and "minutos_jugados" columns have many values which would make same errors in the future, so they will be change.

```python
new_df["MINUTOS_JUGADOS"]=new_df["MINUTOS_JUGADOS"].str.replace("'","")
new_df["RATING"]=new_df["RATING"].replace("-",np.nan)
```

```python
df_final=pd.concat([new_df,columnas,columnas1],axis=1)
```

- Now, the data is clean.

```python
df_final.head()
```

```python
df_final.info()
```

```python
df_final.columns
```

- To get the data with the same type, all the numeric features are going to be float.

```python
nombres=['GOLES', 'ASISTENCIAS', 'MINUTOS_JUGADOS',
       'TIROS_ARCO', 'TIROS_FUERA_ARCO', 'TIROS_BLOQUEADOS','RATING', 
       'DESPEJES', 'TIROS_BLOQUEADOS_POR_EL_JUGADOR', 'INTERCEPCIONES',
       'TACKLES.1', 'DRIBBLED_PAST', 'TOQUES', 'KEY_PASSES',
        'POSSESION_LOST', 'FOULS', 'WAS_FOULED', 'OFFSIDES',
       'SALVADAS', 'PUNCHES', 'HIGH_CLAIMS', 'DUELOS_R', 'DUELOS_E',
       'GAMBETAS_R', 'GAMBETAS_E', 'CROSSES_R', 'CROSSES_E', 'LONG_BALLS_R',
       'LONG_BALLS_E', 'DUELOS_TERRESTRES_R', 'DUELOS_TERRESTRES_E',
       'DUELOS_AEREOS_R', 'DUELOS_AEREROS_E', 'RUNS_OUT_R', 'RUN_OUT_E',
       'PASES_R', 'PASES_E','%PASES_E']
for i in nombres:
    df_final = df_final.astype({i: float})
```
